```python
with open("Corpus_Self guided.txt", "r", encoding="utf-8") as file:
    data=file.read()
```

```python
import nltk
from nltk import word_tokenize
tokenizing = nltk.word_tokenize(data)
tokenized_text = nltk.Text(tokenizing)
```

```python
len(tokenized_text) 
```

```python
removing_punc=[w.lower() for w in tokenized_text if w.isalpha()] # lower case, for loop list comprehension, only taking alphabetical words
```

```python
len(removing_punc)
```

```python
removing_punc[:10]
```

```python
from nltk.corpus import stopwords

removing_stopwords=[]

for word in removing_punc:
    if word not in stopwords.words('english'):
        removing_stopwords.append(word)
```

```python
len(removing_stopwords)
```

```python
removing_stopwords[:]
```

```python
with open('forstylecloud.txt', 'a',encoding='utf-8') as f:
    for i in removing_stopwords:
         f.write(i + ' ')
```

```python
bigram_fd = nltk.FreqDist(nltk.bigrams(removing_stopwords))
bigram_fd.most_common(100)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize= (20,10))
bigram_fd.plot(50)
```

```python
# http://mpqa.cs.pitt.edu/
pos_words = []
neg_words = []
with open("subjclueslen1-HLTEMNLP05.tff") as file:
    for line in file:
        line_attrib = line.split()
        word = line_attrib[2].split('=')[1] 
        polarity = line_attrib[-1].split('=')[1] 
        if polarity =='positive':
            pos_words.append(word)
        elif polarity=='negative':
            neg_words.append(word)
            
print('Total positive words found: ',len(pos_words))
print('Total negative words found: ',len(neg_words))
```

```python
# Text positive and negative words bar korty hoby
pos_word_add = ['impressed', 'innovative', 'standard', 'good', 'unbeatable', 'super', 'advanced', 'beautiful']
for term in pos_word_add:
    pos_words.append(term)

neg_word_add = ['slowly','battery','sometimes', 'charge', 'lower', 'difficult', 'warm', 'problematic', ]
for term in neg_word_add:
    neg_words.append(term)
```

```python
p_t=[]
n_t=[]
for w in removing_stopwords:
    if w in pos_words:
        p_t.append(w)

for w in removing_stopwords:
    if w in neg_words:
        n_t.append(w)
```

```python
text=" ".join(removing_stopwords)
from textblob import TextBlob

blob = TextBlob(text)

sentiment = blob.sentiment.polarity

if sentiment > 0:
    print("The sentiment is positive.")
elif sentiment == 0:
    print("The sentiment is neutral.")
else:
    print("The sentiment is negative.")
```

```python
positive_words_count = len([word for word, tag in blob.tags if TextBlob(word).sentiment.polarity > 0])
negative_words_count = len([word for word, tag in blob.tags if TextBlob(word).sentiment.polarity < 0])
print(positive_words_count)
print(negative_words_count)
```
